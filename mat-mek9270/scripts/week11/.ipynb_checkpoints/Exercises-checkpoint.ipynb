```python
x = sp.Symbol('x')
obj = Helmholtz()
N = 20
u = sp.exp(sp.cos(x-1/2))#*(1-x**2)
alpha = 0.1
uN, uE, xi, l2 = obj(N, alpha, u)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
N_list = np.append(np.arange(6,8),15)
print(N_list)
l2_arr = np.zeros(len(N_list))
UN = []
X= []
ax[0].plot(xi,uE)
for i in range(len(N_list)):
    N = N_list[i]
    uN, uE, xi, l2 = obj(N, alpha, u)   
    l2_arr[i] = l2
    UN.append(uN)
    X.append(xi)
for i in range(len(N_list)):
    ax[0].plot(X[i], UN[i],'--')

ax[1].loglog(N_list,l2_arr)
```
